```
def evaluar_usuario (edad, gusta_musica_urbana):
  if edad >= 18 and gusta_musica_urbana:
    print("bienvenido a la disco")
  elif edad >= 18 and not gusta_musica_urbana:
    print("puedes entrar, pero no se si es lo tuyo")
  elif edad < 18 and gusta_musica_urbana:
    print("no importa tu gusto muscial, no puedes entrar")
  elif edad < 18 and not gusta_musica_urbana:
    print("no puedes entrar y tampoco te gustaria")


evaluar_usuario(15, False)

```
